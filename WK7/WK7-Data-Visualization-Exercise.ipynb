```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("raw/chip-dataset-cleaned.csv",
                 delimiter=",")
df.info()
```

```python
df = df.drop(df.columns[[10,11,12]],axis=1)
df.head()
```

# No. of Chips Manufactured

```python
df_w = df.copy()
```

```python
chip_count = df_w['Vendor'].value_counts()
chip_count
```

```python
type(chip_count)
```

```python
chip_count.plot(kind='bar')
plt.title('Chips Manufactured (CPU+GPU)')
plt.xlabel('')
plt.ylabel('No. of Chips')
plt.show()
```

```python
chart = chip_count.plot(kind='bar',
                        color = ['red','#de6c41','#de6c41','#e49b80','#e49b80'],
                        figsize=(10,5))
chart.bar_label(chart.containers[0], label_type='edge')
plt.title('Chips Manufactured (CPU+GPU)',
          fontfamily = 'Palatino Linotype',
          fontsize=12,
          fontweight='bold')
plt.xlabel('')
chart.set_xticklabels(chip_count.index, rotation=0)
plt.ylabel('No. of Chips')
plt.show()
```

# No. of CPU Manufactured

```python
df_cpu = df_w[df_w['Type']=='CPU']
df_cpu
```

```python
cpu_count = df_cpu['Vendor'].value_counts()
cpu_count
```

# No. of GPU Manufactured

```python
df_gpu = df_w[df_w['Type']=='GPU']
df_gpu
```

```python
gpu_count = df_gpu['Vendor'].value_counts()
gpu_count
```

```python
chart = gpu_count.plot(kind='bar',
                       figsize=(9,3),
                       color = ['red','#de6c41', '#de6c41', '#e49b80', '#e49b80'])
plt.title('GPU Manufactured',
          fontfamily = 'Palatino Linotype',
          fontsize=12,
          fontweight='bold')
plt.ylabel('No. of GPU')
plt.xlabel('')
plt.show()
```

# Lowest TDP (Thermal Design Power)

```python
df_tdp = df_w[['Product','TDP']]
df_tdp
```

```python
type(df_tdp)
```

```python
df_tdp = df_tdp.sort_values(by='TDP')
df_tdp
```

```python
df_tdp = df_tdp.head(7)
df_tdp
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_tdp['Product'],
        df_tdp['TDP'],
        color = ['red','red','red','red','red','red', '#de6c41'])
plt.title('Products with the Lowest TDP',
        fontfamily = 'Palatino Linotype',
          fontsize=12,
          fontweight='bold')
plt.ylabel('Watts')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# GPU Process Size Trend

```python
df_ps = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps
```

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

```python
df_ps = df_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_ps
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_ps['Year'],df_ps['Process_Size'])
plt.title('GPU Process Size Trend',
        fontfamily = 'Palatino Linotype',
          fontsize=12,
          fontweight='bold')
plt.ylabel('nm')
plt.show()
```
